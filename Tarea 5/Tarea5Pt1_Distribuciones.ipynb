**I. Generar 10,000 números aleatorios e ilustrar su distribución para dos casos**

**(a) Distribución uniforme**

```python
#Preguntamos al usuario cuántos números aleatorios quiere generar
print('Cantidad de números aleatorios:', end='')
N = int(input())
```

```python
#Generamos 10,000 números aleatorios con distribución uniforme
sample1 = np.random.uniform(-10,10,N)
```

```python
#Graficamos el histograma de los números aleatorios para observar su distribución
fig = plt.figure()
fig.set_size_inches(5,5)
fig.suptitle('Distribución uniforme de números aleatorios')
ax = fig.add_subplot()
ax.hist(sample1, bins = 1000)
ax.set_xlabel('Valor')
ax.set_ylabel('Frecuencia')
ax.grid()
```

**(b) Distribución Gaussiana**

```python
#Definimos una función de Box-Muller para generar N números aleatorios con distribución gaussiana.
def BoxMullerSample(N):
    sample = np.zeros(N)
    for i in range(N-1):
        #Generamos 2 números aleatorios de la distribución uniforme en el intervalo (0,1)
        U_1 = np.random.uniform()
        U_2 = np.random.uniform()
        #Generamos el nuevo número aleatorio con la condición de Box-Mulle
        sample[i] = np.sqrt(-2*np.log(U_1))*np.cos(2*np.pi*U_2)
    return sample
```

```python
#Preguntamos al usuario cuántos números aleatorios quiere generar
print('Cantidad de números aleatorios:', end='')
N = int(input())
```

```python
#Generamos 10,000 números aleatorios con distribución gaussiana
sample2 = BoxMullerSample(N)
```

```python
#Graficamos el histograma de los números aleatorios para observar su distribución
fig = plt.figure()
fig.set_size_inches(5,5)
fig.suptitle('Distribución gaussiana de números aleatorios (Box-Muller)')
ax = fig.add_subplot()
ax.hist(sample2, bins = 1000)
ax.set_xlabel('Valor')
ax.set_ylabel('Frecuencia')
ax.grid()
```
