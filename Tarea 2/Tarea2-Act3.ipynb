```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Creamos una función que genere un conjunto de puntos que corresponderán a los centros de círculos de un radio dependiente de la concentración reducida y sin traslapes
#Para que no haya traslapes la distancia entre los centros de los círculos debe ser de al menos dos veces el radio
#Utilizaremos el número de partículas y la concentración reducida como argumentos de la función
#Creamos un loop para generar números aleatorios con otro loop dentro que evalue si el punto generado se traslapa con los puntos generados anteriormente
#Si el punto generado traslapa, el punto es descartado
def ranTuples(nr, N):
    i = 0
    muestraX = []
    muestraY = []
    phi = np.pi*nr/4.0 #Se calcula la fracción de área a partir de la concentración reducida
    rad = np.sqrt(phi/np.pi) #Se calcula el radio de las partículas
    L = np.sqrt(N/nr) #Calculamos la longitud de la celda
    while i < N:
        err = False
        x_n = np.random.rand()*L
        y_n = np.random.rand()*L
        if i == 0: #La primer coordenada simplemente la agregamos a la lista
            muestraX.append(x_n)
            muestraY.append(y_n)
            i = i + 1
        elif i > 0: #A partir de la segunda coordenada generada comenzamos a evaluar las distancias entre los puntos
            for j in range(len(muestraX)):
                dist = np.sqrt((x_n - muestraX[j])**2 + (y_n - muestraY[j])**2)
                if dist<2*rad: #Si la distancia es menor a dos veces el radio paramos la iteración e indicamos que hay un error con el valor generado 
                    err = True
                    break
                else:
                    continue
            if err == False: #Si no hay un error registrado entonces agregamos la coordenada generada a la lista
                muestraX.append(x_n)
                muestraY.append(y_n)
                i = i + 1
    muestra = {'x_n': muestraX, 'y_n': muestraY} #Creamos una lista de listas
    return muestra
            
        
```
