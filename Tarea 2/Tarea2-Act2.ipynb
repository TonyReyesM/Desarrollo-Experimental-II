```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#(a) Creamos una muestra de 100 números aleatorios con una semilla seleccionada por el usuario
```

```python
#Creamos una lista del 1 al 100 que se asociará a cada valor de las muestras generadas
nlist = list()
for i in range(1,101):
    nlist.append(i)
```

```python
#Definimos la función que genera la muestra aleatoria de números entre -100 y 100
def ranNum1(sem):
    np.random.seed(sem)
    ranList1 = np.random.rand(100)*200 - 100
    return ranList1
```

```python
#Pedimos al usuario que defina el valor de la semilla
print('Ingrese el valor de la semilla de la muestra aleatoria:', end='')
semU1 = int(input())
```

```python
#Generamos la muestra con la semilla que el usuario ingresó
muestra1 = ranNum1(semU1)
```

```python
#Generamos el dataframe para la muestra y lo exportamos a un archivo de datos
data1 = {'n': nlist, 'x_n': muestra1}
df1 = pd.DataFrame(data1)
df1.to_csv('act2_muestraA.dat', sep =' ', index = False)
```

```python
#(b) Creamos muestras aleatorias de 100, 1000, y 10000 pares con dos semillas diferentes y creamos archivos de datos
```

```python
#Definimos la función para crear las muestras con el número de valores y las semillas como argumentos
#Los números generados estarán entre -1000 y 1000
def ranNum2(sem1, sem2, n):
    np.random.seed(sem1)
    x_n = np.random.rand(n)*2000 - 1000
    np.random.seed(sem2)
    y_n = np.random.rand(n)*2000 - 1000
    data = {'x_n': x_n, 'y_n': y_n}
    return data
```

```python
#Pedimos al usuario que defina el valor de las semillas
print('Ingrese el valor de la primera semilla de la muestra aleatoria:', end='')
semU2_1 = int(input())
print('Ingrese el valor de la segunda semilla de la muestra aleatoria:', end='')
semU2_2 = int(input())
```

```python
#Creamos las muestras para de 100, 1000 y 10000 pares y las colocamos en dataframes
df2_100 = pd.DataFrame(ranNum2(semU2_1, semU2_2, 100))
df2_1000 = pd.DataFrame(ranNum2(semU2_1, semU2_2, 1000))
df2_10000 = pd.DataFrame(ranNum2(semU2_1, semU2_2, 10000))
```

```python
#Creamos archivos de datos para cada muestra
df2_100.to_csv('act2_muestraB1.dat', sep =' ', index = False)
df2_1000.to_csv('act2_muestraB2.dat', sep =' ', index = False)
df2_10000.to_csv('act2_muestraB3.dat', sep =' ', index = False)
```

```python
#(c) Graficamos cada muestra obtenida en el inciso (b)
```

```python
#Graficamos la muestra de 100 pares
plt.scatter(df2_100['x_n'], df2_100['y_n'], c = 'b')
plt.grid()
plt.title('Muestra 1: 100 pares')
plt.xlabel('x_n')
plt.ylabel('y_n')
```

```python
#Graficamos la muestra de 1000 pares
plt.scatter(df2_1000['x_n'], df2_1000['y_n'], c = 'g')
plt.grid()
plt.title('Muestra 2: 1000 pares')
plt.xlabel('x_n')
plt.ylabel('y_n')
```

```python
#Graficamos la muestra de 1000 pares
plt.scatter(df2_10000['x_n'], df2_10000['y_n'], c = 'r')
plt.grid()
plt.title('Muestra 3: 10000 pares')
plt.xlabel('x_n')
plt.ylabel('y_n')
```
